# Puntos equidistantes:

```python
import math
import matplotlib.pyplot as plt

def generate_equidistant_points(num_points):
    # Calcula el número de filas y columnas necesarios para aproximarse a la distancia deseada.
    num_rows_cols = int(math.sqrt(num_points))

    # Asegura que el número de filas y columnas sea igual.
    while num_rows_cols * (num_rows_cols - 1) >= num_points:
        num_rows_cols -= 1

    # Calcula la distancia entre puntos.
    step_size = 1.0 / (num_rows_cols - 1)

    # Genera los puntos equidistantes en la cuadrícula.
    points = []
    for i in range(num_rows_cols):
        for j in range(num_rows_cols):
            x = i * step_size
            y = j * step_size
            points.append((x, y))

    return points

# Número deseado de puntos equidistantes.
num_points = 1000000

# Genera los puntos.
equidistant_points = generate_equidistant_points(num_points)

# Separa los puntos en coordenadas X e Y.
x_values, y_values = zip(*equidistant_points)

# Grafica los puntos.
plt.scatter(x_values, y_values, s=1)  # El parámetro s ajusta el tamaño de los puntos.
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Puntos Equidistantes en una Cuadrícula 1x1')
plt.show()
```

```python
import math

def calculate_distance(point):
    # Calcula la distancia entre un punto y la coordenada (0, 0) utilizando el teorema de Pitágoras.
    return math.sqrt(point[0] ** 2 + point[1] ** 2)

def count_points_within_distance(points, max_distance):
    # Inicializa un contador.
    count = 0

    # Itera a través de los puntos y cuenta aquellos cuya distancia es menor o igual a max_distance.
    for point in points:
        distance = calculate_distance(point)
        if distance <= max_distance:
            count += 1

    return count

# Calcula la distancia máxima deseada (en este caso, 1 o menos).
max_distance = 1.0

# Cuenta los puntos dentro de la distancia deseada.
count = count_points_within_distance(equidistant_points, max_distance)

# Imprime el número de puntos que cumplen con el criterio.
print(f"Número de puntos dentro de una distancia de {max_distance}: {count}")
```

```python
valor = (count / num_points) * 4

print(f"Valor aprox pi: {valor}")
```

# Puntos aleatorios:

```python
import math
import random

def generate_random_points(num_points):
    # Genera puntos aleatorios en el rango [0, 1).
    points = [(random.random(), random.random()) for _ in range(num_points)]

    return points

# Número deseado de puntos aleatorios.
num_points = 1000000

# Genera los puntos aleatorios.
random_points = generate_random_points(num_points)

def calculate_distance(point):
    # Calcula la distancia entre un punto y la coordenada (0, 0) utilizando el teorema de Pitágoras.
    return math.sqrt(point[0] ** 2 + point[1] ** 2)

def count_points_within_distance(points, max_distance):
    # Inicializa un contador.
    count = 0

    # Itera a través de los puntos y cuenta aquellos cuya distancia es menor o igual a max_distance.
    for point in points:
        distance = calculate_distance(point)
        if distance <= max_distance:
            count += 1

    return count

# Separa los puntos en coordenadas X e Y.
x_values, y_values = zip(*random_points)

# Grafica los puntos.
plt.scatter(x_values, y_values, s=1)  # El parámetro s ajusta el tamaño de los puntos.
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Puntos Equidistantes en una Cuadrícula 1x1')
plt.show()
```

```python
# Calcula la distancia máxima deseada (en este caso, 1 o menos).
max_distance = 1.0

# Cuenta los puntos dentro de la distancia deseada.
count = count_points_within_distance(random_points, max_distance)

# Imprime el número de puntos que cumplen con el criterio.
print(f"Número de puntos dentro de una distancia de {max_distance}: {count}")
```

```python
valor = (count / num_points) * 4

print(f"Valor aprox pi: {valor}")
```
