```python
# Librerias
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Cargamos los datos
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [75, 35, 105, 50, 85, 72, 120, 90]
precio = [200, 120, 210, 155, 300, 240, 245, 250]
```

```python
# Convertir lista unidimensional a bidimensional
y = np.array(y).reshape(-1, 1)
precio = np.array(precio).reshape(-1, 1)

#plt.plot(y, precio)
#plt.show()

# Creamos el objeto de Regresión Linear
regr = linear_model.LinearRegression()

# Entrenamos nuestro modelo
regr.fit(y, precio)

# Hacemos las predicciones que en definitiva una línea (en este caso, al ser 2D)
y_pred = regr.predict(y)

# Veamos los coeficienetes obtenidos, En nuestro caso, serán la Tangente
print('Coefficients: \n', regr.coef_)
# Este es el valor donde corta el eje Y (en X=0)
print('Independent term: \n', regr.intercept_)
# Error Cuadrado Medio
print("Mean squared error: %.2f" % mean_squared_error(precio, y_pred))
# Puntaje de Varianza. El mejor puntaje es un 1.0
print('Variance score: %.2f' % r2_score(precio, y_pred))
```

```python
# Visualizar los datos y la recta de regresión
plt.scatter(y, precio, color='blue', label='Datos')
plt.plot(y, y_pred, color='red', linewidth=2, label='Recta de Regresión')
plt.xlabel('y')
plt.ylabel('precio')
plt.legend()
plt.show()
```
